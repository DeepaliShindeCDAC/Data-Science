```python
#5. Explore the shape and soil type through countplot, pie chart
# Countplot for Shape
sns.countplot(x='SHAPE', data=df)
plt.title('Distribution of Shape')
plt.show()

# Pie chart for Shape
df['SHAPE'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Shape')
plt.ylabel('')
plt.show()

# Countplot for Soil Type
sns.countplot(x='SOIL_TYPE', data=df)
plt.title('Distribution of Soil Type')
plt.show()

# Pie chart for Soil Type
df['SOIL_TYPE'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Soil Type')
plt.ylabel('')
plt.show()
```

```python
#Explore alchohol_per, temperature, avg_humidity, fert_nitro_per, water_ph using kdeplot
# List of columns to plot
columns = ['ALCOHOL_PER', 'TEMPERATURE', 'AVG_HUMIDITY', 'FERT_NITRO_PER', 'WATER_PH']

# KDE plots for each variable
for column in columns:
    sns.kdeplot(df[column], shade=True)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
#7. Apply the data encoding (if necessary)
from sklearn.preprocessing import LabelEncoder

# Label encoding
df['SHAPE'] = LabelEncoder().fit_transform(df['SHAPE'])
df['SOIL_TYPE'] = LabelEncoder().fit_transform(df['SOIL_TYPE'])
df
```

```python
# 8. Select only COLOR_INTENSITY FLAVANOIDS PROLINE TEMPERATURE FER_P2O5_PER as input features.
selected_features = ['COLOR_INTENSITY', 'FLAVANOIDS', 'PROLINE', 'TEMPERATURE', 'FER_P2O5_PER']
X_selected = df[selected_features]
X_selected
```

```python
#9. Implement classification algorithm
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y,  random_state=0)
```

```python

# Initialize the classifier
classifier = RandomForestClassifier(random_state=0)
```

```python

# Train the classifier
classifier.fit(X_train, y_train)

```

```python
# Make predictions
y_pred = classifier.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
#10. Show prediction on some unseen data.
```

```python
unseen_data = pd.DataFrame({
    'COLOR_INTENSITY': [5.1, 2.9],
    'FLAVANOIDS': [2.3, 1.7],
    'PROLINE': [800, 650],
    'TEMPERATURE': [25.0, 22.5],
    'FER_P2O5_PER': [15.0, 12.5]
})

# Show prediction on unseen data
unseen_predictions = classifier.predict(unseen_data)
print("Predictions on unseen data:", unseen_predictions)
```
