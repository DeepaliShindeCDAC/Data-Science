### Bar Plot

```python
student=  pd.read_csv('datasets/student.csv')
```

```python
plt.bar(student['name'],student['marks'],color = 'cyan')
```

```python
plt.figure(figsize=(16,9))
plt.bar(student['name'],student['marks'],color = 'cyan')
```

```python
plt.figure(figsize=(16,9))
plt.bar(student['name'],student['marks'],color = 'cyan')
plt.bar(student['name'],student['age'],color = 'red')
```

```python
plt.figure(figsize=(16,9))
plt.bar(student['name'],student['marks'],align='edge', color = 'purple', label ="marks")
plt.bar(student['name'],student['age'],color = 'red', label ="age")
plt.legend()
```

```python
plt.figure(figsize=(16,9))
plt.barh(student['name'],student['marks'], color = 'purple', label ="marks")
plt.barh(student['name'],student['age'],color = 'red', label ="age")
plt.legend()
```
