### 1. Count how many number of posts are paid and unpaid?

```python
mask = f[f['Paid']>0].count()
print('paid posts are:',mask['Paid'])
```

```python
mask = f[f['Paid']==0].count()
print('unpaid posts are:',mask['Paid'])
```

#### 3. Find average number of likes to the post in September month.

```python
gr =f.groupby('Post Month')
gr.groups
```

```python
gr['like'].mean()
```

```python
average_likes_usinggroup = gr.get_group(9)['like'].mean() #using grouping

print("The average number of likes for posts in September is:",average_likes_usinggroup)
```

```python
september_data = f[f['Post Month']== 9]
```

```python
september_data
```

#### 6. In which hour of the day maximum posts are posted?

```python
gr =f.groupby('Post Hour')
gr.groups 
```

```python
posts_per_hour = f['Post Hour'].value_counts()
posts_per_hour
```

```python
max_posts_hour = posts_per_hour.idxmax()
max_posts_count = posts_per_hour.max()
print("The hour of the day with the maximum posts is:", max_posts_hour, "with posts =", max_posts_count)
```

### 7. In which day of October, maximum number of interactions are found?

```python
Oct_data = f[f['Post Month']== 10]
```

```python
Oct_data
```

```python
daywisemaxinteraction = Oct_data.groupby('Post Weekday')['Total Interactions'].max()
 
```

```python
noofmaxinteractions =daywisemaxinteraction.max()
dayofmaxinteractions =daywisemaxinteraction.idxmax();
```

```python
print("maximum number of interactions=",noofmaxinteractions," are found on",dayofmaxinteractions, "weekday")
```

#### 10. Find average and median of interactions for the paid (1) posts.


```python
paid_posts = f[f['Paid'] == 1]
paid_posts

# Calculate the average of interactions for paid posts
average_interactions = paid_posts['Total Interactions'].mean()

# Calculate the median of interactions for paid posts
median_interactions = paid_posts['Total Interactions'].median()

print("average_interactions= ",average_interactions,"and median_interactions=",median_interactions)
```
