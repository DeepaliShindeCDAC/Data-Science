```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('datasets/mtcars.csv')
df
```

```python
x = df[['disp','hp','wt']]
y =df['mpg']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg  =LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.coef_, reg.intercept_
```

```python
reg.score(x,y)
```

```python
disp =221; hp=102; wt=2.91
```

```python
reg.predict([[221,102,2.91]])
```
