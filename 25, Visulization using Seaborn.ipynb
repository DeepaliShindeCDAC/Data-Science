```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
sns.displot(iris['sepal_length'])
```

```python
sns.kdeplot(iris['sepal_length'])
```

```python
sns.histplot(iris['sepal_length'], kde = True)
```

### Joint plot

```python
sns.jointplot(x= 'petal_length', y='petal_width', data = iris)
```

```python
sns.jointplot(x= 'petal_length', y='petal_width', data = iris , hue='species')
```

```python
help(sns.jointplot())
```

```python
sns.jointplot(x= 'petal_length', y='petal_width', data = iris , hue='species' , palette='rocket')
```

### Pair plot

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue ='species')
```

#### Strip plot

```python
sns.stripplot(x='species',y='petal_length',data=iris, hue='species')
```

```python
sns.stripplot(x='species',y='petal_length',data=iris, hue='species',jitter =False)
```

```python
### Swarm plot
```

```python
sns.swarmplot(x='species',y='petal_length',data=iris, hue='species')
```

```python
### Box plot
```

```python
sns.boxplot(x='species',y='petal_length',data=iris, hue='species')
```

### Voilin plot

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.violinplot(x='day',y='total_bill', data =tips , hue="sex")
```

### count plot

```python
sns.countplot(tips["sex"]);
```

```python
help(sns.countplot())
```
